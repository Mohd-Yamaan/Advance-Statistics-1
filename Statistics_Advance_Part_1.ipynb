11. What are the characteristics of a normal distribution?

**Answer --** A normal distribution is a very important probability distribution that looks like a symmetric bell-shaped curve. Here are some key characteristics:

* It is symmetrical around the mean, so both sides of the curve are mirror images.

* The mean, median, and mode are all the same value and located at the center of the distribution.

* Most of the data is concentrated near the mean, and as you move further away, the probabilities get smaller.

* The curve never touches the horizontal axis but gets closer and closer to it (it extends to infinity in both directions).

* About 68% of the data falls within 1 standard deviation from the mean, 95% within 2 standard deviations, and 99.7% within 3 standard deviations ‚Äî this is called the empirical rule.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # Not a valid Bernoulli outcome

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Success
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Failure
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution>

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of the discrete uniform distribution
values = np.arange(1, 11)  # For example, values from 1 to 10
prob = 1 / len(values)     # Equal probability for each value

# Calculate the CDF
cdf = np.cumsum([prob] * len(values))

# Plot the CDF
plt.step(values, cdf, where='post', label='CDF', color='blue')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 10)')
plt.grid(True)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them>


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot the Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='', color='blue')
    plt.axhline(0, color='green', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-Scores of Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-Score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage with random data
sample_data = np.random.normal(50, 10, 100)  # Mean=50, SD=10
z_scores = calculate_and_plot_z_scores(sample_data)
```

18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    z = (x - mean) / std_dev
    return z

def plot_standard_normal_with_z(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)  # Standard normal distribution

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Curve', color='blue')

    # Shade area up to the z-score
    x_shade = np.linspace(-4, z_score, 1000)
    y_shade = norm.pdf(x_shade, 0, 1)
    plt.fill_between(x_shade, y_shade, color='orange', alpha=0.5, label=f'Area ‚â§ Z = {z_score:.2f}')

    # Draw vertical line at Z-score
    plt.axvline(z_score, color='red', linestyle='dashed', label=f'Z = {z_score:.2f}')

    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example input
data_point = 75
mean = 70
std_dev = 5

# Step 1: Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point} is: {z:.2f}")

# Step 2: Plot and compare
plot_standard_normal_with_z(z)
```

19. Implement hypothesis testing using Z-statistics for a sample dataset5


```python
import numpy as np
from scipy.stats import norm

# Step 1: Sample and Population Info
sample_mean = 98          # xÃÑ
population_mean = 100     # Œº (claimed mean)
sample_std = 10           # s
sample_size = 50          # n
alpha = 0.05              # significance level

# Step 2: Calculate Z-statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
print(f"Z-statistic: {z:.2f}")

# Step 3: Calculate critical value for 2-tailed test
z_critical = norm.ppf(1 - alpha / 2)
print(f"Z-critical (¬±): ¬±{z_critical:.2f}")

# Step 4: Decision
if abs(z) > z_critical:
    print("Reject the null hypothesis: The claim may be incorrect.")
else:
    print("Fail to reject the null hypothesis: The claim seems reasonable.")
```

20. Create a confidence interval for a dataset using Python and interpret the result5


```python
import numpy as np
import scipy.stats as stats

# Sample data (example: exam scores of 10 students)
data = [65, 70, 72, 68, 75, 80, 78, 85, 90, 88]

# Step 1: Calculate mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)  # standard error of the mean

# Step 2: Set confidence level (let‚Äôs use 95%)
confidence = 0.95

# Step 3: Calculate confidence interval
ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=std_err)

# Step 4: Print results
print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")
```

21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.


```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for consistent results
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate the mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)

# Step 3: Set confidence level (e.g., 95%) and calculate CI
confidence = 0.95
ci = stats.t.interval(confidence, df=len(data)-1, loc=mean, scale=std_err)

# Step 4: Print results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")
```

23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
